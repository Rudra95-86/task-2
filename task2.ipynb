```python
import pandas as pd
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
sns.histplot(data['Age'].dropna(),kde=True)
plt.title("Age Distribution")

plt.subplot(2,2,2)
sns.boxplot(x="Survived",y="Age",data=data)
plt.title("Age Distribution by survival")

plt.subplot(2,2,3)
sns.histplot(data['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')

plt.subplot(2,2,4)
sns.boxplot(x='Survived', y='Fare', data=data)
plt.title('Fare Distribution by Survival')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,14))
numerical_features = ["Age","Fare","SibSp","Parch","Survived"]
sns.pairplot(data[numerical_features].dropna(),hue='Survived')
plt.suptitle('Pairplot of Numerical Feature by Survival',y=1.0)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.show()

data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
plt.figure(figsize=(10, 5))
sns.barplot(x='FamilySize', y='Survived', data=data)
plt.title('Survival Rate by Family Size')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='Embarked', y='Survived', data=data)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```

```python
print("\n Key Statistics:")
print(f"Overall survival rate: {data['Survived'].mean()*100:.2f}%")
print("\nSurvival rate by class:")
print(data.groupby('Pclass')['Survived'].mean())
print("\nSurvival rate by gender:")
print(data.groupby('Sex')['Survived'].mean())
print("\nAverage fare by class:")
print(data.groupby('Pclass')['Fare'].mean())
print("\nAverage age by survival:")
print(data.groupby('Survived')['Age'].mean())
```
